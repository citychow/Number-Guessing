```python
#prompt the users to input a range of numbers

import random
import matplotlib.pyplot as plt

#print("Please enter your name: ")
#player1=input()
#print("Hello "+str(player1)+", let's play the number guessing game!")

#prompt the user to select a range of value
print("Please select the range of number:")
print("Smaller no.: ")
s=input()
print("Larger no.:")
l=input()
print("Selected range of number is "+str(s)+" to "+str(l))

def game():
    #generate a random no.
    ans=random.randint(int(s), int(l))
    #print("The no. in my mind is: "+str(ans))
    #prompt the user to make the guess
    print("Make a guess! ")
    i=1
    j=1
    trail=i
    listoftrail=[]
    gameplay=[]
    guess=int(input())
    while(guess!=ans): #use while loop instead of for loop
        if guess<int(s) or guess>int(l):
            print("The no. is out of range. Try again.")
            guess=int(input())
        if guess > ans:
            print("Give me a smaller no.")
            i=i+1
            listoftrail.append([i])
            guess=int(input())
        if guess < ans:
            print("Give me a larger no.")
            i=i+1
            listoftrail.append([i])
            guess=int(input())
    print("BINGO! finally. You've tried " +str(i) +" times. Try again?")
    playagain=str(input())
    if playagain=='yes':
        j=j+1
        print(j)
        gameplay.append([j])
        game()
    if playagain!='yes':
        print("Goodbye. You've played "+str(j)+" times.")
    plt.plot(gameplay, listoftrail,'.r')
game()
```
